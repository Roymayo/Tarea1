```python
import numpy as np
import time as time
import matplotlib.pyplot as plt
#La documentación se realizó con la ayuda de la IA DeepSeek
```

```python
def ProductoPuntoBloques(X, Y):
    """
    ProductoPuntoBloques(X, Y) -> float o str

    Esta función calcula el producto punto (o producto escalar) de dos vectores X e Y
    particionando los vectores en bloques de una sola coordenada. Ambos vectores deben
    tener la misma cantidad de elementos. Si no es así, la función devuelve un mensaje
    de error.

    Parámetros:
    -----------
    X : list
        El primer vector, representado como una lista de números.
    Y : list
        El segundo vector, representado como una lista de números.

    Retorna:
    --------
    float o str
        - Si los vectores son del mismo tamaño, la función devuelve el producto punto
          de los vectores X e Y.
        - Si los vectores no son del mismo tamaño, la función devuelve una cadena
          indicando que los vectores deben tener la misma cantidad de elementos.

    Ejemplo:
    --------
    >>> X = [1, 2, 3]
    >>> Y = [4, 5, 6]
    >>> ProductoPuntoBloques(X, Y)
    32

    >>> X = [1, 2, 3]
    >>> Y = [4, 5]
    >>> ProductoPuntoBloques(X, Y)
    'Los vectores deben tener la misma cantidad de elementos'
    """
    m = len(X)
    n = len(Y)
    PP = 0
    if m == n:
        for i in range(m):
            SubX = [X[i]]  # Particiona el vector X en su i-ésima coordenada
            SubY = [Y[i]]  # Particiona el vector Y en su i-ésima coordenada
            PP += SubX[0] * SubY[0]  # Multiplica las coordenadas y acumula el resultado
    else:
        return 'Los vectores deben tener la misma cantidad de elementos'
    return PP
```

```python
def ProductoPuntoVectores(X, Y):
    """
    ProductoPuntoVectores(X, Y) -> float o str

    Esta función calcula el producto punto (o producto escalar) de dos vectores X e Y.
    Ambos vectores deben tener la misma cantidad de elementos. Si no es así, la función
    devuelve un mensaje de error.

    Parámetros:
    -----------
    X : list
        El primer vector, representado como una lista de números.
    Y : list
        El segundo vector, representado como una lista de números.

    Retorna:
    --------
    float o str
        - Si los vectores son del mismo tamaño, la función devuelve el producto punto
          de los vectores X e Y.
        - Si los vectores no son del mismo tamaño, la función devuelve una cadena
          indicando que los vectores deben tener el mismo tamaño.

    Ejemplo:
    --------
    >>> X = [1, 2, 3]
    >>> Y = [4, 5, 6]
    >>> ProductoPuntoVectores(X, Y)
    32

    >>> X = [1, 2, 3]
    >>> Y = [4, 5]
    >>> ProductoPuntoVectores(X, Y)
    'Los vectores deben tener el mismo tamaño'
    """
    m = len(X)
    n = len(Y)
    PP = 0
    if m == n:
        for i in range(m):
            PP += X[i] * Y[i]  # Multiplica las coordenadas correspondientes y acumula el resultado
    else:
        return 'Los vectores deben tener el mismo tamaño'
    return PP
```

```python
#Celda con el código que compara el producto punto obtenido de manera normal y por bloques
Tamaño = np.zeros(90) #Almacena la dimensión de los vectores, con n = 50,...,140
Tiempo1 = np.zeros(90) #Almacena el tiempo de ejecución para el producto punto por bloques
Tiempo2 = np.zeros(90) #Almacena el tiempo de ejecución para el producto punto normal
for i in range(90):
    Tamaño[i] = i+50 #Llena Tamaño
    #Crea los vectores que se utilizarán para el producto punto con números uniformemente aleatorios 
    A = np.random.uniform(1,100,(i+50,i+50))
    B = np.random.uniform(1,100,(i+50,i+50))
    #Calcula el tiempo de ejecución del método por bloques
    inicio = time.time()
    ProductoPuntoBloques(A,B)
    fin = time.time()
    tiempo = fin-inicio
    Tiempo1[i] = tiempo #Llena Tiempo1
    #Calcula el tiempo de ejecución del método normal
    inicio2 = time.time()
    ProductoPuntoVectores(A,B)
    fin2 = time.time()
    tiempo2 = fin2-inicio2
    Tiempo2[i] = tiempo2 #Llena Tiempo2
```

```python
#Grafica el tamaño del vector vs el tiempo de ejecución del producto punto por ambos métodos 
plt.plot(Tamaño,Tiempo1,label='PPBloques')
plt.xlabel('Tamaño')
plt.ylabel('Tiempo')
plt.legend()
plt.plot(Tamaño,Tiempo2,label='PPVectores')
plt.legend()
plt.title('Tamaño de la matriz vs tiempo de ejecución para el producto punto')
plt.show()
```
